##  키오스크 

### 1 메뉴 출력
- 음료와 가격을 보여줍니다.

### 2 주문 기능
- 메뉴 번호를 입력받습니다.
- HOT / ICE 선택이 가능합니다.

### 3 결제 기능
- 카드 / 현금 선택
- 현금 시 거스름돈 계산

### 4 주문 내역
- 주문한 메뉴와 총 금액 출력

```python
class Kiosk:
    def __init__(self):
        self.menu = ['아메리카노', '바닐라라떼', '모카라떼', 'yuza_tea', 'green_tea', 'choco_latte']
        self.price = [3000, 3500, 4000, 5000, 4500, 5000]
        self.order_menu = []
        self.order_price = []
        self.price_sum = 0

    # 메뉴 출력
    def menu_print(self):
        for i in range(len(self.menu)):
            print(i+1, self.menu[i], ':', self.price[i], '원')

    # 주문 선택
    def menu_select(self):
        n = 0
        while n < 1 or n > len(self.menu):
            try:
                n = int(input("음료 번호를 입력하세요 : "))
                if 1 <= n <= len(self.menu):
                    self.order_menu.append(self.menu[n-1])
                    self.order_price.append(self.price[n-1])
                    self.price_sum += self.price[n-1]
                else:
                    print("없는 메뉴입니다.")
            except ValueError:
                print("숫자를 입력하세요.")

        # 온도 선택
        while True:
            try:
                t = int(input("HOT은 1, ICE는 2 입력 : "))
                if t == 1:
                    self.temp = "HOT"
                    break
                elif t == 2:
                    self.temp = "ICE"
                    break
                else:
                    print("1 또는 2 입력")
            except ValueError:
                print("숫자를 입력하세요.")

        print(self.temp, self.menu[n-1], ':', self.price[n-1], '원')

    #결제수단 선택하기
    def pay(self):
        
        while True:
            method = input("결제 수단 (cash/1, card/2) : ").lower()
            if method == 'cash' or method == '1':
                print("직원을 호출하겠습니다.")
                break
            elif method == 'card' or method == '2':
                print("IC칩 방향에 맞게 카드를 꽂아주세요.")
                break
            else:
                print("다시 입력하세요.")

        # 금액 입력
        while True:
            try:
                money = int(input("금액 입력 : "))
                if money < self.price_sum:
                    print("금액 부족")
                else:
                    print("결제 완료")
                    print("거스름돈 :", money - self.price_sum, "원")  # 개선점 카드를 입력해도 거스름돈이 뜨는건 모순이므로 고쳐봐야겠음
                    break
            except ValueError:
                print("숫자 입력")

    # 주문서 출력
    def table(self):
        print("\n[주문 내역]")
        for i in range(len(self.order_menu)):
            print(self.order_menu[i], ':', self.order_price[i], '원')

        print("총 금액 :", self.price_sum, "원")
```

```python
k = Kiosk()
k.menu_print()
k.menu_select()
k.pay()
k.table()
```
